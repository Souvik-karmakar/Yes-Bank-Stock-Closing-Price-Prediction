We can see that some amount of outliers are present in the dataset , Lets investigate further for those outlies.

````python
#Function to detect outliers
def outlier_find(s_price):
  q1=s_price.quantile(0.25)
  q3=s_price.quantile(0.75)
  IQR=q3-q1
  outliers= s_price[((s_price<(q1-1.5*IQR)) | (s_price>(q3+1.5*IQR)))]
  return outliers
````

````python
#find outlier count
cols=['Open','High','Low','Close']
for i in cols:
  outliers = outlier_find(s_price[i]).count()
  print(f'{i} : {outliers}')
````

````python
#find outliers
cols=['Open','High','Low','Close']
for i in cols:
  outliers = outlier_find(s_price[i])
  print(f'{i} : {outliers}')
````

#### Chart - 6 Seasonal Decomposition

````python
# Chart - 6 visualization code
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(s_price['Close'], model='additive', period=12)
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(result.observed)
plt.title('Observed')
plt.subplot(412)
plt.plot(result.trend)
plt.title('Trend')
plt.subplot(413)
plt.plot(result.seasonal)
plt.title('Seasonal')
plt.subplot(414)
plt.plot(result.resid)
plt.title('Residual')
plt.tight_layout()
plt.show()
````

Answer:- The specific chart chosen is the seasonal decomposition plot, which decomposes the Yes Bank stock closing prices into four components: observed, trend, seasonal, and residual. This chart was chosen to analyze the underlying patterns in the data over time, including any seasonality present.

Answer:-

**Alternative Hypothesis (H1):** The mean closing price of Yes Bank stock is not equal to the historical average closing price.

#### 2. Statistical test:- One-sample t-test

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_1samp
# Assuming historical average closing price of Yes Bank stock is 100
t_stat, p_value = ttest_1samp(s_price['Close'], 100)
print("One-sample t-test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Answer:- To obtain P-Value i have done One Sample T-Test.

##### Why did you choose the specific statistical test?

Answer:- We choose the one-sample t-test because it allows us to test whether the mean closing price of Yes Bank stock significantly differs from a specific value (in this case, the historical average closing price). The null hypothesis assumes that there is no difference between the mean closing price and the historical average, while the alternative hypothesis suggests otherwise.As the P-value (0.4736123563972122) is greater than predetermined significance level (e.g., 0.05), we failed to reject the null hypothesis and conclude that The mean closing price of Yes Bank stock is equal to the historical average closing price.








### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer:-

**Null Hypothesis (H0)**: The closing prices of Yes Bank stock are non-stationary, indicating the presence of a unit root in the time series data.

**Alternate Hypothesis (H1)**: The closing prices of Yes Bank stock are stationary, suggesting the absence of a unit root in the time series data.

#### 2. Statistical test-ADF test

````python
# Perform Statistical Test to obtain P-Value
from statsmodels.tsa.stattools import adfuller

# Perform ADF test
result = adfuller(s_price['Close'])

# Extract ADF test statistic and p-value
adf_statistic = result[0]
p_value = result[1]

# Print results
print("ADF test:")
print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)

# Interpret results
if p_value < 0.05:
    print("Reject the null hypothesis (H0): The closing prices are stationary.")
else:
    print("Fail to reject the null hypothesis (H0): The closing prices are non-stationary.")
````

##### Which statistical test have you done to obtain P-Value?

Answer:- To obtain the P-Value we have done ADF test.

##### Why did you choose the specific statistical test?

Answer:- We chose the Augmented Dickey-Fuller (ADF) test for hypothesis testing because it's specifically designed to determine whether a time series is stationary or non-stationary. For the given hypothesis, we are interested in assessing whether the closing prices of Yes Bank stock exhibit stationarity over time. As the P-value is greater than 0.05 we can conclude that the stock closing price is non-stationary.

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Define hyperparameters for grid search
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize Random Forest Regression model
model_rf = RandomForestRegressor(random_state=42)

# Perform Grid Search Cross-Validation
grid_search_rf = GridSearchCV(model_rf, param_grid_rf, cv=5, scoring='neg_mean_squared_error')
grid_search_rf.fit(X_train, y_train)

# Best hyperparameters
best_params_rf = grid_search_rf.best_params_

# Initialize Random Forest Regression model with best hyperparameters
model_rf_best = RandomForestRegressor(**best_params_rf, random_state=42)

# Fit the model to the training data
model_rf_best.fit(X_train, y_train)

# Predict on the test set
y_pred_rf_best = model_rf_best.predict(X_test)

# Evaluation metrics
mse_rf_best = mean_squared_error(y_test, y_pred_rf_best)
r2_rf_best = r2_score(y_test, y_pred_rf_best)

print("Best Hyperparameters for Random Forest Regression:", best_params_rf)
print("Mean Squared Error (MSE) for Random Forest Regression with Hyperparameter Tuning:", mse_rf_best)
print("R-squared (R2) Score for Random Forest Regression with Hyperparameter Tuning:", r2_rf_best)
````

**Model Used: Linear Regression**

**Linear Regression** is a statistical technique used for modeling the relationship between a dependent variable and one or more independent variables. In the context of stock price prediction, Linear Regression assumes a linear relationship between the input features (such as historical stock prices, volume, etc.) and the target variable (closing price).

Linear Regression fits a linear equation to the observed data by minimizing the sum of the squares of the differences between the observed and predicted values. The resulting model equation can be expressed as:


```
y=b0 + b1x1 + b2x2 +....+ bnxn

y= predicted closing price,
b0= is the intercept term,
b1, b2,....,bn are the coefficients of the input features x1,x2,....,xn respectively
```
**Feature Importance Using Permutation Importance:**

Permutation Importance measures the increase in the model's error (e.g., MSE) when the values of a feature are randomly shuffled while keeping other features unchanged. The greater the increase in error after shuffling a feature, the more important that feature is for the model's predictions.



````python
from sklearn.inspection import permutation_importance

# Fit the Random Forest model
model_lr.fit(X_train, y_train)

# Calculate permutation importance
perm_importance = permutation_importance(model_lr, X_test, y_test, n_repeats=30, random_state=42)

# Get feature names
feature_names = X.columns

# Display feature importance scores
print("Permutation Importance Scores:")
for i, feature in enumerate(feature_names):
    print(f"{feature}: {perm_importance.importances_mean[i]}")
````
